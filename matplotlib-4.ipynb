## Object Oriented Interface
Though matplotlib.pyplot offers a bunch of features to plot and visualize the data, that are easy to use, it is recommended to use Object Oriented Interface using matplotlib.axes.Axes class. It is highly encouraged way to use matplotlib as it gives more power and control to manipulate the plots.
matplotlib objects are defined below, take a look before diving straight into the code part:
* FigureCanvas-> Container class for a figure instance
* Figure-> Container for axes instances(can be one or more
* Axes -> an area to hold basic elements such as lines, text etc.

```python
# simple example code to use OO approach
fig = plt.figure()
x = np.arange(10)
y = x**2

# The dimensions are as [left, bottom, width, height] of the new axes.
# All quantities are in fractions of figure width and height.
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel("X")
ax.set_ylabel("X**2")
ax.plot(x,y)
plt.show()
```

Now, in the above code, the function plt.figure() returns a figure having a default size(if not assigned any) where one or more axes instances can be added and the function fig.add_axes(), creates and returns an axes instance where we can do the plotting.
(Axes is basically the area where your plot appears).

### Subplots
Just like the function add_axes returns an axes instance, add_subplot()s not just returns an axes instance but also adds the subplot, i.e., it adds an Axes to the figure as part of a subplot arrangement.
Now add_subplot() function requires 3 parameters:
* nrows -> number of rows of subplots to be prepared
* ncolumnns -> number of columns of subplots to be prepared
* index -> specifies the axes of the subplot we are working on currently

```python
# example to create subplots using add_subplots
fig = plt.figure(figsize=(8,8))
x = np.arange(10)
y = x**2

x1 = np.arange(10)
y1 = x**3
# here add_subplot() function takes three arguements(by defailt are (1,1,1)),
# here ,2 specifies-> 2 rows, then 1 specifies-> 1 Colummn and then 1 specifies-> the first subplot that we have prepared
ax1 = fig.add_subplot(2,1,1)
ax1.plot(x,y)
ax1.set_title("Subplot-1")
ax1.set_xlabel("x")
ax1.set_ylabel("x**2")

# here ,2 specifies-> 2 rows, then 1 specifies-> 1 Colummn and then 2 specifies-> the second subplot that we have prepared
# we can also change the facecolor in order to distinguish between the subplots
ax2 = fig.add_subplot(2,1,2,facecolor = 'c')
ax2.plot(x1,y1)
ax2.set_xlabel("x")
ax2.set_ylabel("x**3")
ax2.set_title("Subplot-2")

plt.show()
```

```python
# using the above example to show that add_subplot function does not overwrite the existing
# this example is just to show some additional features, we rarely use it
fig = plt.figure(figsize=(8,8))
x = np.arange(10)
y = np.exp(-x)

x1 = np.arange(10)
y1 = x**3
# here add_subplot() function takes three arguements(by defailt are (1,1,1)),
# here ,2 specifies-> 2 rows, then 1 specifies-> 1 Colummn and then 1 specifies-> the first subplot that we have prepared
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x,y)
ax1.set_title("Subplot-1")
ax1.set_xlabel("x")
ax1.set_ylabel("exp**(-x)")

# here ,2 specifies-> 2 rows, then 1 specifies-> 1 Colummn and then 2 specifies-> the second subplot that we have prepared
# we can also change the facecolor in order to distinguish between the subplots
ax2 = fig.add_subplot(2,2,2,facecolor = 'c')
ax2.plot(x1,y1)
ax2.set_xlabel("x")
ax2.set_ylabel("x**3")
ax2.set_title("Subplot-2")

plt.show()
```

Another way to create subplots is by using the matplotlib.pyplot.subplots function. 
This is almost identical to add_subplot functionality but I consider it to be more robust and easy to handle rather than
add_subplot function.
This function takes in two major arguements defined below:
* nrows-> number of rows of subplots to be prepared(default is 1)
* ncolumns-> number of columns of subplots to be prepared(default is 1)

and this function even returns a tuple as (fig,ax) that are defined below:
* fig-> subplots return a plt.figure object to be used as a container for all the subplots, stored in variable fig
* ax -> subplots returns an object of axes.Axes class and returns an array of objects in case of multiple plots, stored in variable ax 

```python
# example to create plots using plt.subplots function

fig,ax = plt.subplots(2,2)

x = np.arange(10)
y = x**2

x1 = np.arange(10)
y1 = 2*x

x2 = np.linspace(-3,3,10)
y2 = x**3

x3 = np.linspace(-1,1,10)
y3 = np.sin(x)

ax[0,0].plot(x,y)# row=0, col=0
ax[0,1].plot(x1,y1,color='y') # row=0, col=1
ax[1,0].plot(x2,y2,color='r') # row=1,col=0
ax[1,1].plot(x3,y3,color='g') # row=1, col=1


ax[0,1].set_facecolor("cyan")
ax[1,0].set_facecolor("yellow")

plt.show()
```

Some other parameters of plt.subplots are:
* sharex, sharey: Specifies sharing of properties between axes. Possible values are none, all, row, col or a boolean with a default value of False.

```python
# example to show share axes, that is all the plots having a common shared axis

# figure having a shape of 3,1 i.e. 3 rows and 1 column 
fig,ax = plt.subplots(3,1,sharex='all')
x = np.linspace(-3,3,10)

ax[0].plot(x,x)
ax[1].plot(x*2,x*2,color='g')
ax[2].plot(x*3,x*3,color='r')

plt.show()
```

### Adding additional X or Y axis
We can even add an additional axis in a single plot using the function ax.twinx(), let's understand it with an example

```python
# example to show twinx function usage
fig,ax = plt.subplots(1,1)
x = np.linspace(0,5,20)
y1 = x**2
y2 = -1*(x**2)

ax.plot(x,y1)
ax.set_ylabel("y1-axis")

ax_ = ax.twinx()

ax_.plot(x,y2,color="m")
ax_.set_ylabel("y2-axis")

# this function below is used to set the xticks as
ax.set_xticks([0,1,2,3,4,5])

# there is also a function to set the xticks and yticks labels which is called as ax.set_xticklabels/set_yticklabels
# we'll use it here only just for the explanation
ax.set_xticklabels(["zero","one","two","three","four","five"])

plt.show()
```

### Changing to logarithmic scale
Sometimes one or more data points has values that are much larger and cannot be shown properly in a normal scale, and therefore we change the scale to logarithmic scale. we can change the scale by using the functions set_xscale() and set_yscale()

```python
# example to change to logarithmic scale

fig,ax = plt.subplots(1,2)
x = np.arange(5)

ax[0].plot(x,10**x)
ax[0].set_title("Normal Scale")

ax[1].set_yscale("log")
ax[1].plot(x,10**x,color="r")
ax[1].set_title("Logarithmic Scale")

plt.show()
```

## Contour Plots
+ Contour plot is the graphical technique for representing a 3-dimensional surface on a 2-dimensional format.
+ Contour pllot is appropriate if you want to check how some value z changes as a function of two inputs x and y.(z=f(x,y))
+ It graphs two variables x y and a response variable z as contour. These contours are sometimes called as iso-response values.
+ Contour plot is formed by two independent variables i.e., vertical axis and horizontal axis and one dependent variable i.e., iso-response values.

So, basically any type of data in which you want to see the change in a value of a variable based on the changes in another two independent variables, you can go for a contour plot.

Now, matplotlib provides us with two functions that are
* contour() -> for contour lines
* contourf() -> to fill contours
Note-> function signatures and return values are same for both the functions

some important parameters are discussed below:
+ X,Y -> Two independent variables, that is, coordinates of the values in z. Both X and Y must be 2-d with same shape as z created via np.meshgrid() function
+ Z -> Height values over which the contour is drawn
+ levels -> this is array like input that determines the number and positions of the contour lines / regions.
+ cmap -> A Colormap instance or registered colormap name. The colormap maps the level values to colors.
+ norm -> If a colormap is used, the Normalize instance scales the level values to the canonical colormap range [0, 1] for mapping to colors.
and it returns a Quadcontourset that creates and stores a set of all colour lines or filled regions.

```python
# example that show how to graph a contour plot that creates contour lines

x = np.linspace(0,5,100)
y = np.linspace(0,5,100)

X,Y = np.meshgrid(x,y)
# We used the np.meshgrid() function to get the coordinate matrices, X and Y
# from the two coordinate vectors, x and y.

Z = np.sqrt(X**2+Y**2)

fig,ax = plt.subplots(1,1,figsize=(10,8))

cont = ax.contour(X,Y,Z)

ax.set_title('Contour Plot')

plt.show()
```

```python
# example that show how to graph a contourf plot

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)
# We used the np.meshgrid() function to get the coordinate matrices, X and Y
# from the two coordinate vectors, x and y.

Z = X**2+Y**2

fig,ax = plt.subplots(1,1,figsize=(10,8))

# here we first created the countours just like the above example gave color to those contour as red in this case
contour = ax.contour(X,Y,Z,colors="r")

# after creating the countour we filled the contours using countour filled function
contour_filled = ax.contourf(X,Y,Z)

# adding a color bar
fig.colorbar(contour_filled) 
ax.set_title('Contour Plot')

plt.show()
```
