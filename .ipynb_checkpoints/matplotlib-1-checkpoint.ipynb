## What is Matplotlib?
Matplotlib is a plotting library in python, that is used to create high quality graphs,charts and figures. It is a library that covers all the minute details of a figure. 
It helps in data visualization which makes it easier for us to understand the data well. 
In this notebook we'll be covering the basics of matplotlib library as we know it.

We mostly use the module matplotlib.pyplot when it comes to plotting the graphs. 
It is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

Now, let us dive in and plot our graphs in order to understand how it all works.

```python
import matplotlib.pyplot as plt
import numpy as np
```

As you can see in the above block of code,
we have imported matplotlib.pyplot as plt which means we'll be referencing pyplot with plt in the entire notebook.
P.S. -> It's a general norm to create an alias of module pyplot as plt so you don't have to write the library name on every cell.

## Line plots
These type of plots are the simplest yet one of the most important plots when it comes to matplotlib.
The most basic way to plot a graph is using plt.plot() function which plots the graph and then plt.show() which displays it.

```python
plt.plot([2,4,6,8])
plt.show()
```

Well, you might be wondering How can we plot a single array without any other variable on a 2-d graph ? How is it possible ?,
well in that case if we do not mention any other array then by default [0,1,2,3] is assumed to be an X input and the values entered by you are assumed to be Y input. Similarly for an array length 5 the default values will be [0,1,2,3,4] and so on. 

Let us understand this with an example as well-

```python
plt.plot([3,6,9,12,15])
```

Here, we can clearly see that the X-axis is having default values as [0,1,2,3,4], but another issue is, How can we plot a graph without using the show() function?

Well in that case, this is an IPython notebook in which we are running each and every line one by one, so, the IPython interpreter knows that we want it to plot a graph even without using the show() function, But when we use any other python IDE(like pycharm), we always have to use this show() function after defining each and every plot so that the interpreter knows that we want it to show a particular plot with the above defined values. Therefore, from now on we'll use plt.show() for every single plot.

From the above 2 examples, We can see how easy it was to plot a graph with this library.
In this we'll be moving forward focusing on the examples and visualizations rather than theory

## Adding labels and title

```python
# This is a simple line plot that takes two values of x and y.
x=[1,2,3,4,5]
y=[2,5,3,2,6]
plt.plot(x,y)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Simple Line plot")
plt.show()
```

The above plot is also a line plot between two variables x and y. It uses ylabel and xlabel are used to set the label text for y and x coordinates and plt.title function is used to set the title of the plot.

```python
# plotting x and x**2 graph
x = np.arange(1,10)
plt.plot(x,[y**2 for y in x])
plt.xlabel("x")
plt.ylabel("x**2")
plt.title("Square plot")
plt.show()
```

## Multiline Plots

```python
# we can even plot multiline graphs as
x = np.arange(1,7)

# for (x,x)
plt.plot(x,x)

# for (x,x**2)
plt.plot(x,[y**2 for y in x])

# for (x,x**3)
plt.plot(x,[y**3 for y in x])

plt.show()
```

## Defining figure and adding Legend

```python
plt.figure(figsize=(8,8))
x = np.arange(1,7)

# for (x,x)
plt.plot(x,x, label="Linear")

# for (x,x**2)
plt.plot(x,[y**2 for y in x], label="Quadtratic")

# for (x,x**3)
plt.plot(x,[y**3 for y in x], label="Cubic")

plt.legend()
plt.show()
```

In the above plot we've first created the figure or the skeleton of the plot on our own and have mention the size as(8,8),
if we don't create it using plt.figure then it creates a default skeleton of it's own in order to display the plot.

And, another function that we've used is plt.legend(), this function displays the labels of all the line plots along with colour of line, on the graph itself. But to use legend functionality we first have to label all the plots differently as we've done it in the above example.

This way we can make our plots more intuitive.

```python
# multiline plot for sine and cosine waves
plt.figure(figsize=(10,7))
x = np.linspace(0,10,100)

# sine plot
plt.plot(np.sin(x),label="Sine Wave")

# cosine plot
plt.plot(np.cos(x),label="Cosine Wave")

plt.legend()
plt.show()
```

## Grids and Axes

### Grids

While using matplotlib, there are situations where if we have a reference system we can improve our understanding of the plot.
For this we can use grids, by simply calling the grid function which takes only 2 values i.e, True to enable it or False to disable it.

```python
# example to use grid
x = np.linspace(-5,5,100)

# for (x,x**2)
plt.plot(x,x**3)

# for (x,e**x), here 'e' is exponential function
plt.plot(x,np.exp(x))

plt.grid(True)
plt.show()
```

### Axes

Matplotlib automatically sets the scale of the graph precisely according to the data. However, it gives us the functionality to change the scale of the graph according to us.
In this we'll define the scale of the graph by using the function plt.axis()

```python
# let us use the above example for this

x = np.linspace(-5,5,100)

# for (x,x**2)
plt.plot(x,x**3)

# for (x,e**x), here 'e' is exponential function
plt.plot(x,np.exp(x))

plt.axis() # this will show us the current axes of the graph
```

```python
x = np.linspace(-5,5,100)

# for (x,x**2)
plt.plot(x,x**3)

# for (x,e**x), here 'e' is exponential function
plt.plot(x,np.exp(x))
plt.axis([-7,10,-170,170]) # this will change the axes of the graph
plt.show()
```

### Final Plot

```python
plt.figure(figsize=(9,9))
x = np.linspace(-3,3,50)

# for (x,x)
plt.plot(x,x, label="Linear")

# for (x,x**2)
plt.plot(x,x**2, label="Quadtratic")

# for (x,x**3)
plt.plot(x,x**3, label="Cubic")

# for (x,e**x), here 'e' is exponential function
plt.plot(x,np.exp(x), label="Exponential")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Combined Plot")

plt.axis([-5,5,-40,40])
plt.legend()
plt.grid(True)
plt.show()
```
