## -Bar Charts
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.
we use bar() function to plot a bar graph in matplotlib.

```python
# plotting a simple bar graph
x = np.arange(5)
y = np.random.randint(20,size=5)

plt.bar(x,y)
plt.show()
```

Some important parameters to the bar function are:
* bottom - represents bottom coordinates of the bar
* height - represents height of the bar
* width - represents width of the bar, width of the bar graph is 0.8 by default
* align - centre or edge, default is centre
* color - to set the color of the bar

```python
# let's take an example now to plot a comparison bar graph with random values
# to know about the percentage of students studying python,java and cpp in previous years.
# In this graph we are comparing three values(Python,java,cpp)

plt.figure(figsize=(8,8))

x = np.arange(5)
y1 = np.random.randint(100,size=5)
y2 = np.random.randint(100,size=5)
y3 = np.random.randint(100,size=5)

width = 0.27 # here we have set the length of the width

plt.bar(x,y1,color="b",width=width, label='C++')
plt.bar(x+width,y2,color="y",width=width, label='Java')
plt.bar(x+2*width,y3,color="c",width=width, label = 'Python')

plt.xticks(range(len(x)),["2016","2017","2018","2019","2020"])

plt.legend()
plt.show()
```

In the above example, we have assigned value of 0.27 to width variable and then have added it in the x parameter in the bar plot. This has been done because by default the value of the width parameter is 0.8 but here we have changed it to 0.27, and later on have added this value to the green bar and twice of this value to the red bar so that they all the bars for a particular year are plotted together without spaces between them.

Value added to the graphs is similar to that of width so that they stick together. And value of y is initialized using random function and is different for all the three plots.

### Stacked Up Bar Graph
A stacked bar graph (or stacked bar chart)  is a chart that uses bars to show comparisons between categories of data, but with ability to break down and compare parts of a whole. Each bar in the chart represents a whole, and segments in the bar represent different parts or categories of that whole.

To plot this type of graph we just use the bottom parameter in order to stack one value over the other.
Let's understand this one with an example

```python
# for this we'll be using an example showing total number of men and women in a class each year, where the whole bar
# will represent the total number of students in a class

year = [2016, 2017, 2018, 2019,2020]  
men =  [20,32,41,24,36]  
women = [31,32,34,18,23]

# the first call is as usual
plt.bar(year, men, color="y",width=0.35)

# the second one is special to create stacked bar plots
plt.bar(year, women, bottom=men, color="g",width=0.35)

plt.title("Stacked-bar")
plt.xlabel('Year')  
plt.ylabel('students')

plt.show()  
```

## -Pie Charts
A type of graph in which a circle is divided into sectors(also called wedges) that each represent a proportion of the whole.
matplotlib provides us a function called as pie() to plot pie-charts in python.

Note - The fractional area of each sector is given by x/sum(x). 
So, if sum(x)< 1, then no normalization is done and pie chart is drawn using x values directly, which results in discontinuity of the cart.

```python
# this is a simple example of pie chart to randomly show the usage of social media handles.
# Note-> this is just an example based on random numbers not actual data
plt.figure(figsize=(7,7))

x = [50,30,45]
plt.pie(x,labels=["instagram","facebook","twitter"])

plt.legend()
plt.show()
```

Important parameters to the pie function are given below: 
* labels -> a sequence of string providing a label for each sector/wedge
* colors -> a list of Matplotlib colors, cyclically used starting from the assigned color, to color the sectors/wedges
* explode -> If specified, it's an array of the same length as that of X. Each of its values specify the radius fraction with which to offset the sector/wedge from the center of the pie.
* shadow -> It draws the shadow of the sector/wedge of the pie
* autopct -> String, used to label the sectors/wedges with their numeric values. The labels are placed inside the sectors

```python
# example which uses all the parameters of the pie chart
# in this also we'll be plotting random values for usage of different social media files
plt.figure(figsize=(10,10))
x = np.random.randint(100,size=5)

labels = ["whatsapp","instagram","facebook","snapchat","twitter"]

explode = [0,0.2,0,0,0.1]

# here autopct has this fixed format only the 1.1f here is variable and here it means value upto one decimal place
plt.pie(x,labels=labels,explode=explode,autopct='%1.1f%%',shadow=True)

plt.legend()
plt.show()
```

## - Histogram Charts
They are the graphical representations of the distribution of numerical data, represented as bars. It gives the probability distribution of a continuous variable.
matplotlib provides us with a function called as hist() to plot histograms in python.

```python
# example to plot a simple histogram
x = np.random.randn(100)

plt.hist(x)
# here by default the value of bins=10 , i.e., only 10 bars are computed within the given data limit by default

plt.show()
```

Histogram parameters are given below:
* bins -> no. of bars to be shown, integer or sequence, by default value is 10.
* range -> lower and upper range of the bins.
* cumulative -> If True, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values.
* histtype -> histtype is the type of histogram. These are- 1.bar 2.barstacked 3.step 4.stepfilled

```python
# examples to show uses of different parameters

x = np.random.randint(150,size=25)

plt.hist(x,bins=15)

plt.show()
```

```python
x = np.random.randint(150,size=25)

plt.hist(x,bins=15,cumulative=True,histtype="step")

plt.show()
```

## - Boxplots
+ The box plot (also known as box and whisker diagram) is a standardized way of displaying the distribution of data based on the five number summary: minimum, first quartile, median, third quartile, and maximum.
+ In the simplest box plot the central rectangle spans the first quartile to the third quartile.
+ The line in the middle of the rectangle represents the median of the distribution
+ The whiskers at both the ends of the box plots represents the lower extreme and upper extreme of the distribution.

To get a better understanding look at the image I found on google given below:


Matplotlib provides us with a function name plt.boxplot() in order to plot a boxplot

```python
# example to plot the boxplots
# here we have created fake data with np.random.normal()
#It takes three arguments, mean and standard deviation of the normal distribution, and the number of values desired.
data_1 = np.random.normal(125,7,100)

plt.boxplot(data_1)

plt.show()
```

## - Scatter Plots
+ Scatter plots are one of the most commonly used data visualization plot, just like line plots. In this data points are represented individually by a circle,dot or any other shape instead of being joined by line segments.
+ A scatter plot is often used to identify potential association between two variables(x and y), and it's often drawn before working on a fitting regression function. It gives a good visual picture of the correlation, in particular for nonlinear relationships.


matplotlib provides us with a function scatter() to display scatter plots

```python
# example to easily grasp the concept

x = np.random.randn(1000)
y = np.random.randn(1000)
                      
plt.scatter(x,y)                
plt.show()
```

Some of the parameters of scatter plots are:
* s -> this is used to change the size of the marker, it can be any single value or an array of same size as that of x and y.
* c -> it represents the color of the points in the plot, it can also be any single value or an array or same size as that of x and y
* marker -> it specifies the types of markers used to represent the data points. These markers have  certain abbreviations that are listed below in the table. 

| Marker Value |  Description |
| -----------------  |  ----------- |
|s |Square|
|o |Circle|
|^ |Triangle up|
|v |Triangle down|
|> | Triangle right|
|> |Triangle right|
|< |Triangle left|
|d |Diamond|
|p |Pentagon|
|h |Hexagon|
|8 |Octagon|
|+ |Plus|
|x |Cross|

```python
# example that uses all the parameters mentioned above
plt.figure(figsize=(9,7))

x = np.random.randn(1000)
y = np.random.randn(1000)

size = np.random.randn(1000)
color = np.random.rand(1000)

plt.title("Scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.scatter(x,y,s=200*size,c=color, marker='8')                
plt.show()
```
